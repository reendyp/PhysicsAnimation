## 45 Degrees vs Everydegrees

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from matplotlib.animation import FFMpegWriter
plt.rcParams['animation.ffmpeg_path'] = 'C:\\ffmpeg\\ffmpeg.exe'
```

```python
# Initiation Parameters

g = 9.81
vI = 60.0
q = np.pi/4
y0 = 0.0
```

```python
# Function
def time(q):
    return (vI * np.sin(q) + np.sqrt((vI * np.sin(q))**2 + 2 * g * y0)) / g

def xPos(q,t):
    return vI*t*np.cos(q)
def yPos(q,t):
    return y0 + vI*np.sin(q)*t - 0.5*g*t**2
#yPos(q,t)
```

```python
xList = [0]
yList = [y0]
xList0 = [0]
yList0 = [y0]
xList1 = [0]
yList1 = [y0]
xList2 = [0]
yList2 = [y0]

timeList = np.linspace(0,time(np.pi/4),300)
timeList0 = np.linspace(0,time(np.pi/3),300)
timeList1 = np.linspace(0,time(np.pi/6),300)
timeList2 = np.linspace(0,time(np.pi/12),300)

def calculated():
    for t in timeList:
        xList.append(xPos(q,t))
        yList.append(yPos(q,t))
    for t in timeList0:
        xList0.append(xPos(np.pi/3,t))
        yList0.append(yPos(np.pi/3,t))
    for t in timeList1:
        xList1.append(xPos(np.pi/6,t))
        yList1.append(yPos(np.pi/6,t))
    for t in timeList2:
        xList2.append(xPos(np.pi/12,t))
        yList2.append(yPos(np.pi/12,t))
calculated()
```

```python
# Technical Prosedures to Produce Animation
fig, ax = plt.subplots()

route0, = ax.plot(xList0[0],yList0[0],label='60 Derajat',color='r')
route, = ax.plot(xList[0],yList[0],label='45 Derajat',color='b')
route1, = ax.plot(xList1[0],yList1[0],label='30 Derajat',color='m')
route2, = ax.plot(xList2[0],yList2[0],label='15 Derajat',color='g')


routedot0, = ax.plot(xList0[0],yList0[0],'bo')
routedot, = ax.plot(xList[0],yList[0],'ro')
routedot1, = ax.plot(xList1[0],yList1[0],'ko')
routedot2, = ax.plot(xList2[0],yList2[0],'yo')


ax.set_xlim(0,max(xList)+10)
ax.set_ylim(0,max(yList0)+10)
ax.set_xlabel('Jarak (m)')
ax.set_ylabel('Ketinggian (m)')
ax.legend()
ax.grid()

def animateIt(i):
    route0.set_data(xList0[:i+1],yList0[:i+1])
    routedot0.set_data(xList0[i],yList0[i])
    
    route1.set_data(xList1[:i+1],yList1[:i+1])
    routedot1.set_data(xList1[i],yList1[i])
    
    route2.set_data(xList2[:i+1],yList2[:i+1])
    routedot2.set_data(xList2[i],yList2[i])
    
    route.set_data(xList[:i+1],yList[:i+1])
    routedot.set_data(xList[i],yList[i])
ani = anim.FuncAnimation(fig,animateIt,frames=len(timeList0))

ffmep_writer = anim.FFMpegWriter(fps=30)
ani.save('casey0equal060vs45.mp4',writer=ffmep_writer)
```

## $Y_0$ $!=$ 0

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from matplotlib.animation import FFMpegWriter

plt.rcParams['animation.ffmpeg_path'] = 'C:\\ffmpeg\\ffmpeg.exe'

# Parameter Awal
g = 9.81
vI = 60.0
q = 0.0
y0 = 40.0

# Fungsi Waktu Total Peluncuran
def time_of_flight():
    return (vI * np.sin(q) + np.sqrt((vI * np.sin(q))**2 + 2 * g * y0)) / g

# Fungsi Posisi X dan Y
def xPos(t):
    return vI * t * np.cos(q)

def yPos(t):
    return y0 + vI * t * np.sin(q) - 0.5 * g * t**2

# Menghitung Waktu Total dan Daftar Waktu
total_time = time_of_flight()
timeList = np.linspace(0, total_time, 300)

# Menghitung Posisi
xList = [xPos(t) for t in timeList]
yList = [yPos(t) for t in timeList]

# Pengaturan Grafik
fig, ax = plt.subplots()
route, = ax.plot([], [], 'b')
routedot, = ax.plot([], [], 'ro')

ax.set_xlim(0, max(xList) + 10)
ax.set_ylim(0, max(yList) + 10)
ax.set_xlabel('Jarak (m)')
ax.set_ylabel('Ketinggian (m)')
ax.grid()

# Fungsi Animasi
def animateIt(i):
    route.set_data(xList[:i+1], yList[:i+1])
    routedot.set_data(xList[i], yList[i])
    return route, routedot

ani = anim.FuncAnimation(fig, animateIt, frames=len(timeList), interval=20, blit=True)

# Menyimpan Animasi
ffmep_writer = FFMpegWriter(fps=30)
ani.save('casey0withZeroDegrees.mp4', writer=ffmep_writer)
```
